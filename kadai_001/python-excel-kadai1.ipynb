```python
import openpyxl

# openpyxl.stylesモジュールのPatternFillクラスをインポート
from openpyxl.styles import PatternFill

# openpyxl.stylesモジュールのFontクラスをインポート
from openpyxl.styles import Font

# openpyxl.stylesモジュールのBorderクラスとSideクラスをインポート
from openpyxl.styles import Border, Side

# datetimeモジュールのdatetimeクラスをインポートする
from datetime import datetime

# 新規ワークブックを作成
workbook = openpyxl.Workbook()

# アクティブなシートを取得する
ws = workbook.active

# セルに値を書き込む
ws['B2'].value = "請求書"

ws['F4'].value = "NO."
ws['G4'].value = "0001"

ws['F5'].value = "日付"
today = datetime.today()
yyyymmdd = today.strftime('%Y/%m/%d')
ws['G5'].value = yyyymmdd

company = [
    ["株式会社ABC"],
    ["〒101-0022 東京都千代田区神田練塀町300"],
    ["TEL:03-1234-5678 FAX:03-1234-5678"],
    ["担当者名:鈴木一郎 様"]
]

company_start_row = 4 #４行目
company_start_col = 2 #B列

for i, row in enumerate(company):
    cell = ws.cell(row=company_start_row + i, column=company_start_col)
    cell.value = row[0]

#表を作成
data = [
    ['商品名','数量','単価', "金額"],
    ['商品A',2 ,10000, 20000 ],
    ['商品B', 1, 15000, 15000],
]

data_start_row = 10  # 10行目
data_start_col = 2  # B列

for i, row in enumerate(data):
    for j, value in enumerate(row):
        ws.cell(row=data_start_row + i, column=data_start_col + j).value = value

# 計算
ws['B15'].value = "小計"
ws['B16'].value = "消費税"
ws['B17'].value = "合計"

syokei = data[1][3] + data[2][3]
tax = syokei * 0.1
total = syokei + tax

ws['E13'] = syokei
ws['E15'] = syokei
ws['E16'] = tax
ws['E17'] = total

# Excelファイルを保存
seikyuusyo = f"請求書_{today.strftime('%Y%m%d')}.xlsx"

workbook.save(seikyuusyo)
```
