```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import PatternFill

# Excelファイルを読み込む
df_2022 = pd.read_excel("2022_年間売上表.xlsx")
df_2023 = pd.read_excel("2023_年間売上表.xlsx")

# 縦に結合
total_sales = pd.concat([df_2022, df_2023], axis=0)

# 売上年、商品ごとに金額（千円）の合計を集約
result = total_sales.groupby(['商品', '売上年']).agg({'金額（千円）': 'sum'}).reset_index()

# Excelファイルを作成する
output_file = pd.ExcelWriter('売上集計表.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
result.to_excel(output_file, sheet_name='Sheet1', index=False)

# Excelファイルを保存する
output_file.close()

# 出力したExcelファイルを編集してタイトル行を薄いグレーにする
wb = load_workbook(output_file)
ws = wb.active
header_fill = PatternFill(start_color="F2F2F2", end_color="F2F2F2", fill_type="solid")
for cell in ws[1]:  # 1行目（タイトル行）を対象
    cell.fill = header_fill
wb.save(output_file)

# 結果を表示
print("売上集計表を保存しました")
```
